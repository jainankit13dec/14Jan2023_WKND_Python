```python
import pandas as pd
import numpy  as np
```

```python
data = {"name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],      
       "score":[90,80,85,75,95,60,65],      
       "sport":["Wrestling","Football","Skiing","Swimming","Tennis",
               "Karete","Surfing"],      
       "sex":["M","M","M","M","F","F","F"]}
```

```python
type(data)
```
