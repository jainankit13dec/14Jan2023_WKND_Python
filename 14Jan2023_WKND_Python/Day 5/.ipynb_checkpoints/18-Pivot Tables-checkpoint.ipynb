# Pivot Tables

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {"class":list("ABC")*4,
     "lesson":["math","stat"]*6,
     "sex":list("MFMM")*3,
     "sibling":[1,2,3]*4,
     "score":np.arange(40,100,5)})
```

```python
df.head(10)
```

```python
df.groupby("lesson")["score"].mean()
```

```python
df.groupby(
    ["lesson",
     "class"])[
    "score"].aggregate("mean").unstack()
```

```python
df.head(10)
```

```python
df.pivot_table(
    "score",
    index="lesson",
    columns="class")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",margins=True)
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",fill_value=0)
```
