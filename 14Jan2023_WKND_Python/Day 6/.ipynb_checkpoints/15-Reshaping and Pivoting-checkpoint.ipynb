## Pivoting “Long” to “Wide” Format

```python
import pandas as pd, numpy as np
```

```python
stock=pd.DataFrame(
    {"fruit": ["apple", "plum","grape"]*2,
     "color": ["purple","yellow"]*3,
     "piece":[3,4,5,6,1,2]})
```

```python
stock
```

```python
stock.pivot("fruit", "color", "piece")
```

```python
stock["value"]=np.random.randn(len(stock))
```

```python
stock
```

```python
p=stock.pivot("fruit","color")
p
```

```python
p["value"]
```
