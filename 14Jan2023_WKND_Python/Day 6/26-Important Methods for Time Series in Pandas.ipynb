# Important Methods for Time Series in Pandas

```python
import pandas as pd 
import numpy as np
```

## resampling method

```python
fb=pd.read_csv("Datasets/FB.csv", 
               parse_dates=["Date"], 
               index_col="Date")
```

```python
fb.head()
```

```python
fb.resample("M").mean()
```

```python
%matplotlib inline
```

```python
fb.Close.resample("M").mean().plot()
```

```python
fb.Close.resample("Q").mean()
```

## Shifting

```python
fb1=pd.DataFrame(fb.Close["2019-03"])
```

```python
fb1.head()
```

```python
fb1.shift(1)
```

```python
fb1.shift(-1)
```

```python
fb1["Previous Price"]=fb1.shift(1)
```

```python
fb1
```

```python
fb1["One Day Difference"]=fb1["Close"]-fb1["Previous Price"]
```

```python
fb1.head()
```

```python
fb1["Percentage Change"]=(
    fb1["Close"]-fb1[
        "Previous Price"])*100/fb1[
    "Previous Price"]

fb1
```

```python
fb2=fb1[["Close"]]
```

```python
fb2.head()
```

```python
fb2.index
```

```python
fb2.index=pd.date_range(
    "2019-03-01", periods=21, freq="B")
```

```python
fb2.index
```

```python
fb2.tshift(1)
```

```python
fb2.tshift(-2)
```
